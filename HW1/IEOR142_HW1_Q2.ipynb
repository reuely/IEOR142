Question 2a

```python
accord_training_data = accord[(accord['Year'] >= 2014) & (accord['Year'] <= 2018)][['AccordSales', 'Unemployment', 'AccordQueries', 'CPIEnergy','CPIAll', 'MilesTraveled']]
accord_testing_data = accord[(accord['Year'] >= 2019) & (accord['Year'] <= 2023)][['AccordSales','Unemployment', 'AccordQueries', 'CPIEnergy','CPIAll', 'MilesTraveled']]
display(accord_training_data.head(), accord_testing_data.head())
```

```python
train, test = train_test_split(accord_training_data, test_size = .15, random_state = 42)
display(train.head(), test.head())
```

```python
ols1 = smf.ols(formula= 'AccordSales ~ Unemployment + AccordQueries + CPIEnergy + CPIAll + MilesTraveled', data = train)
model1 = ols1.fit()
print(model1.summary())
```

```python
def VIF(df, columns):
  values = sm.add_constant(df[columns]).values
  num_columns = len(columns)+1
  vif = [variance_inflation_factor(values, i) for i in range(num_columns)]
  return pd.Series(vif[1:], index=columns)
cols = ['Unemployment', 'AccordQueries', 'CPIEnergy', 'CPIAll', 'MilesTraveled']
VIF(train, cols)
```

```python
ols2 = smf.ols(formula= 'AccordSales ~ Unemployment + AccordQueries + CPIEnergy + MilesTraveled', data = train)
model2 = ols2.fit()
print(model2.summary())
```

```python
cols = ['Unemployment', 'AccordQueries', 'CPIEnergy', 'MilesTraveled']
VIF(train, cols)
```

```python
ols3 = smf.ols(formula= 'AccordSales ~ AccordQueries + CPIEnergy + CPIAll', data = train)
model3 = ols3.fit()
print(model3.summary())
```

```python
cols = ['AccordQueries', 'CPIEnergy', 'CPIAll', 'AccordSales']
VIF(train, cols)
```

```python
def OSR2(model, df_train, df_test, dependent_var):
  y_test = df_test[dependent_var]
  y_pred = model.predict(df_test)
  SSE = np.sum((y_test - y_pred)**2)
  SST = np.sum((y_test - np.mean(df_train[dependent_var]))**2)
  return 1 - SSE/SST
```

```python
 OSR2(model3, train, test, 'AccordSales')
```

```python
 OSR2(model3, accord_training_data, accord_testing_data, 'AccordSales')
```
